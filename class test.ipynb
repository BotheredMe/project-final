```python
import nltk
import csv
import pandas as pd
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tokenize import PunktSentenceTokenizer
from nltk.corpus import stopwords
import nltk.classify.util
from Models import worddata
```

```python
lyric_df= pd.read_csv("combined_lyrics.csv")
```

```python
lyric_df
```

```python
wordcounts = []
nostopwordcounts = []
positivewordcounts = []
negativewordcounts = []
neutralwordcounts = []


for data in lyric_df["lyrics"]:
    words = worddata.word_calcs(data)
    wordcount = words.wordcount()
    wordcounts.append(wordcount)
    stopwordcount = words.nostopwordcount()
    nostopwordcounts.append(stopwordcount)
    positivewordcount = words.positivewordcount()
    positivewordcounts.append(positivewordcount)
    negativewordcount = words.negativewordcount()
    negativewordcounts.append(negativewordcount)
    neutralwordcount = words.neutralwordcount()
    neutralwordcounts.append(neutralwordcount)
    
    
```

```python
lyric_df["wordcount"] = wordcounts
lyric_df["nostopwordcount"] = nostopwordcounts
lyric_df["positivewordcount"] = positivewordcounts
lyric_df["negativewordcount"] = negativewordcounts
lyric_df["neutralwordcount"] = neutralwordcounts
```

```python
lyric_df = lyric_df[["genre","artist","song","lyrics","wordcount", "nostopwordcount","positivewordcount","negativewordcount","neutralwordcount"]]
```

```python
lyric_df.to_csv('lyrics.csv')
```
