```python
#dependencies
import nltk
import csv
import pandas as pd
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tokenize import PunktSentenceTokenizer
from nltk.corpus import stopwords
import nltk.classify.util
from Models import worddata
```

```python
# Read in csv of webscraped data
lyric_df= pd.read_csv("combined_lyrics.csv")
```

```python
lyric_df
```

```python
# set up arrays that will be appended through the loop of current data
wordcounts = []
nostopwordcounts = []
positivewordcounts = []
positivepercents = []
negativewordcounts = []
negativepercents = []
neutralwordcounts = []
neutralpercents = []


    
    
```

```python
# loop through data
for data in lyric_df["lyrics"]:
    # Initiate class that will tokenize, count the words
    words = worddata.word_calcs(data)
    wordcount = words.wordcount()
    wordcounts.append(wordcount)
    stopwordcount = words.nostopwordcount()
    nostopwordcounts.append(stopwordcount)
    positivewordcount = words.positivewordcount()
    positivewordcounts.append(positivewordcount)
    # Checking to avoid dividing by 0
    if positivewordcount == 0:
        positivepercents.append(0.00)
    else:
        positivepercent = round(positivewordcount/stopwordcount*100,2)
        positivepercents.append(positivepercent)
    negativewordcount = words.negativewordcount()
    negativewordcounts.append(negativewordcount)
    if negativewordcount == 0:
        negativepercents.append(0.00)
    else:
        negativepercent = round(negativewordcount/stopwordcount*100,2)
        negativepercents.append(negativepercent)
    neutralwordcount = words.neutralwordcount()
    neutralwordcounts.append(neutralwordcount)
    if neutralwordcount == 0:
        neutralpercents.append(0.00)
    else:
        neutralpercent = round(neutralwordcount/stopwordcount*100,2)
        neutralpercents.append(neutralpercent)

    
```

```python
# Input newly calculated arrays as columns in the dataframe

lyric_df["wordcount"] = wordcounts
lyric_df["nostopwordcount"] = nostopwordcounts
lyric_df["positivewordcount"] = positivewordcounts
lyric_df["positivepercents"]=positivepercents
lyric_df["negativewordcount"] = negativewordcounts
lyric_df["negativepercents"]=negativepercents
lyric_df["neutralwordcount"] = neutralwordcounts
lyric_df["neutralpercents"]=neutralpercents
```

```python
# Remove unnecessary columns
lyric_df = lyric_df[["genre","artist","song","lyrics","wordcount", "nostopwordcount","positivewordcount","positivepercents","negativewordcount","negativepercents","neutralwordcount","neutralpercents"]]
```

```python
# Saving as new csv
lyric_df.to_csv('lyrics.csv')
```
